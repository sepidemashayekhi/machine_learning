## Importing the libraries

```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
import sklearn 
```

## Load and Prepare Data

```python
data=pd.read_csv('Churn.csv')
```

## EDA

```python
def EDA(data):
    print('info data:',data.info(),'\n')
    print ("*data shape* ",data.shape)
    print ("data columns",data.columns)
    print ( "Null val",data.isnull().sum())
    return data.head()
EDA(data)
```

## Data Preprocessing

```python
data.drop(['RowNumber','CustomerId','Surname','CreditScore'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['Geography']=le.fit_transform(data['Geography'])
data['Gender']=le.fit_transform(data['Gender'])
```

```python
pd.crosstab(data.Gender,data.Exited,normalize='index')*100
```

```python
pd.crosstab(data.Geography,data.Exited,normalize='columns')*100
```

## Strorytelling - Visualization

```python
plt.figure(figsize=(7,7))
plt.pie(
    data['Exited'].value_counts(sort=True)
    , labels=[ 'Stayed','Churned']
    , autopct='%1.1f%%'
    , textprops={'fontsize':14}
)

plt.show()
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(
    data.corr()
    , annot=True
)
```

```python
sns.pairplot(data, hue="Exited")
```

```python
data.head()
```

## Prepare Data for Machine learning

```python
x=data.iloc[:,0:9].values
y=data.iloc[:,-1].values
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

## Train your model

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(x_train,y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV
param=[{'n_neighbors':np.arange(1,50)}]
graid_search=GridSearchCV(estimator=knn,param_grid=param,scoring='accuracy',cv=10)
graid_search.fit(x_train,y_train)
```

```python
graid_search.best_params_
```

```python
graid_search.best_score_
```

```python
best_knn=KNeighborsClassifier(n_neighbors=9)
best_knn.fit(x_train,y_train)
```

```python
from sklearn.naive_bayes import GaussianNB
bayes=GaussianNB()
bayes.fit(x_train,y_train)
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
```

```python
param_svc=[{'C':[0.25,0.5,0.75,1,1.5,2,3,4],'kernel':['linear']}
          ,{'C':[0.25,0.5,0.75,1,1.5,2,3,4],'kernel':['rbf']}]
grid_svc=GridSearchCV(estimator=svc,param_grid=param_svc,scoring='accuracy',cv=10,n_jobs=-1)
```

```python
grid_svc.fit(x_train,y_train)
```

```python
grid_svc.best_params_
```

```python
best_svc=SVC(C=0.75,kernel='rbf')
best_svc.fit(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
tree.fit(x_train,y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest=RandomForestClassifier()
random_forest.fit(x_train,y_train)
```

```python
from xgboost import XGBClassifier
xg=XGBClassifier()
xg.fit(x_train,y_train)
```

## Test the model and show the metrics

```python
from sklearn.metrics import accuracy_score
```

```python
y_log=log.predict(x_test)
ac_log=accuracy_score(y_test,y_log)
ac_log
```

```python
y_knn=best_knn.predict(x_test)
ac_knn=accuracy_score(y_test,y_knn)
ac_knn
```

```python
y_bayes=bayes.predict(x_test)
ac_bayes=accuracy_score(y_test,y_bayes)
ac_bayes
```

```python
y_best_svc=best_svc.predict(x_test)
ac_best_svc=accuracy_score(y_test,y_best_svc)
ac_best_svc
```

```python
y_random_forest=random_forest.predict(x_test)
ac_random_forest=accuracy_score(y_test,y_random_forest)
ac_random_forest
```

```python
y_xg=xg.predict(x_test)
ac_xg=accuracy_score(y_test,y_xg)
ac_xg
```
